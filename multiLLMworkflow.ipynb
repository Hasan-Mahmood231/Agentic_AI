```python
from langgraph.graph import StateGraph , START , END
from typing import TypedDict
```

```python
from langchain_ollama import ChatOllama

# Initialize the actual chat model object
model = ChatOllama(model="qwen2.5:3b")
```

```python
class blogllm(TypedDict):
    titile : str 
    output : str
```

```python
def Generate_outline(state : blogllm) -> blogllm:
    titile = state['titile']

    prompt = f"generate the 3 short outline of the {titile} for blog creation."

    output = model.invoke(prompt)
    
    state['output'] = output

    return state
```

```python
def generate_blog(state : blogllm) -> blogllm:
    question = state['output']

    prompt = f"create the very short blog on {question}"

    output = model.invoke(prompt)

    state['output'] = output.content

    return state
```

```python
graph = StateGraph(blogllm)
graph.add_node("generate_outline",Generate_outline)
graph.add_node("generate_blog",generate_blog)
```

```python
graph.add_edge(START,"generate_outline")
graph.add_edge("generate_outline","generate_blog")
graph.add_edge("generate_blog",END)

workflow = graph.compile()
```

```python
question = {"query":"create blog on crypto market"}
final_answer = workflow.invoke(question)
print(final_answer)
```
